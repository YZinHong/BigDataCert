- round(dfnew['minmax_scale'].mean(), 4): 최소-최대 척도 변환 후 평균
- dfnew['zscore_scale']=(x-x.mean(axis=0)/x.std(axis=0): Z-score 변환

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
iris = load_iris()
irisdata = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
print(irisdata.head())
print('petal length의 (평균+표준편차)')
print(irisdata['petal length (cm)'].mean() + irisdata['petal length (cm)'].std())
print(irisdata.describe())
print(irisdata.describe()['petal length (cm)']['mean'] + irisdata.describe()['petal length (cm)']['std'])
print('petal width의 (평균+표준편차)')
print(irisdata['petal width (cm)'].mean() + irisdata['petal width (cm)'].std())
dfnew = irisdata[irisdata['target'] == 0]
print(dfnew.head())
x = dfnew['petal length (cm)']
dfnew['minmax_scale'] = (x-x.min(axis=0)/(x.max(axis=0)-x.min(axis=0)))
print('최소-최대 척도 변환 후 petal length의 평균')
print(round(dfnew['minmax_scale'].mean(), 4))
print('Z-score 변환 후 petal length의 평균')
dfnew['zscore_scale'] = (x-x.mean(axis=0)/x.std(axis=0))
print(round(dfnew['zscore_scale'].mean(), 20))
print(dfnew.head())
print(dfnew.describe())
```

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
air = pd.read_csv('/content/drive/MyDrive/work/airquality.csv', index_col=0)
print(air.head())
print(air.isnull().sum())
data = air.dropna(axis=0)
print(data.head())
print(data.isnull().sum())
print('++++++++++++++++++++')
q1 = data['Ozone'].quantile(q=0.25)
q2 = data['Ozone'].quantile(q=0.5)
q3 = data['Ozone'].quantile(q=0.75)
print('Ozone의 1사분위수: ', end=''); print(q1)
print('Ozone의 2사분위수: ', end=''); print(q2)
print('Ozone의 3사분위수: ', end=''); print(q3)
print('^^^^^^^^^^^^^^^^^^^^')
print('Ozone의 평균(데이터 대체전): ', end=''); print(data['Ozone'].mean())
print('!!!!!lambda expressions!!!!!')
nam = lambda x: 0 if ((x>=q3) | (x<=q1)) else x
data['Ozone'] = data['Ozone'].apply(nam)
print('Ozone의 평균(데이터 대체후): ', end=''); print(data['Ozone'].mean())
print('Ozone의 평균+표준편차(데이터 대체후): ', end=''); print(round(data['Ozone'].mean() + data['Ozone'].std(), 5))
```
