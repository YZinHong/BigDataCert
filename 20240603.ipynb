```python
def Get_temperature_days(x, y):
  N = len(x)
  low_days = 0
  high_days = 0
  for i in range(0, N):
    if(x[i] >= y):
      high_days = high_days + 1
    else:
      low_days = low_days + 1
  return [high_days, low_days]

temperature = [18.5, 24.1, 24.0, 22.1, 24.3, 20.2, 20.8, 24.6, 18.2, 20.0, 24.8]
print(Get_temperature_days(temperature, 24))

new_temperature = [24.9, 18.8, 25.2, 22.7, 30.5, 19.4, 20.6, 23.9, 26.2, 17.8, 14.6]
print(Get_temperature_days(new_temperature, 24))
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from importlib import reload
plt = reload(plt)

d = list(range(1,11,1))
print(d)
x=pd.DataFrame(d, columns=['number'])
print(x)
lam = lambda t: t**2
y = x.pipe(lam)
print(y)

plt.plot(x, y, label='y=x**2')
plt.xlabel('x value')
plt.ylabel('y=x**2')
plt.title('y=x**2 Graph')
plt.legend()
plt.show()
```
